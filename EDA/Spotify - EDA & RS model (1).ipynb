## Model Building

```python
#popu = pd.DataFrame(data.groupby('album')['popularity'].mean())
#popu.head()
```

```python
popu = pd.DataFrame(data.groupby('name')['popularity'].mean())
popu.head()
```

```python
song_matrix = data.pivot_table(index = 'album', columns = 'name', values ='popularity')
song_matrix.head()
```

```python
song_matrix_2 = song_matrix.fillna(0)
song_matrix_2
```

```python
popu.sort_values('popularity', ascending = False).head(10)
```

```python
abc = song_matrix_2["Resurrender - Live"]
```

```python
similar_to_abc = song_matrix_2.corrwith(abc)
similar_to_abc
```

```python
corr_abc = pd.DataFrame(similar_to_abc , columns=['corr'])
corr_abc.dropna(inplace=True)
corr_abc.head()
```

```python
corr_abc_2 = corr_abc.join(popu['popularity'])
corr_abc_2
```

```python
corr_abc_2[corr_abc_2['popularity']>50].sort_values(by='corr', ascending = False)
```
